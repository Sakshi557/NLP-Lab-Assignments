```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
from nltk.corpus import wordnet as wn

word = "home"

# Get all synsets of the word
synsets = wn.synsets(word)

# Print synsets
print("Synsets:")
for s in synsets[:3]:
    print("-", s.name(), ":", s.definition())

# Synonyms
synonyms = set()
for s in synsets:
    for lemma in s.lemmas():
        synonyms.add(lemma.name())

print("\nSynonyms:", synonyms)

# Antonyms
antonyms = set()
for s in synsets:
    for lemma in s.lemmas():
        if lemma.antonyms():
            antonyms.add(lemma.antonyms()[0].name())

print("\nAntonyms:", antonyms)

# Hypernyms (general terms)
hypernyms = synsets[0].hypernyms()
print("\nHypernyms:")
for h in hypernyms:
    print("-", h.name(), ":", h.definition())
```
