```python
import re
import nltk
import pandas as pd

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
texts = [
    "I love Natural Language Processing",
    "NLP is very interesting",
    "I am learning machine learning"
]

labels = ["Positive", "Positive", "Neutral"]
```

```python
#text cleaning
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)   # remove punctuation & numbers
    return text
```

```python
#lemmatization and stopword removal

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess(text):
    words = text.split()
    words = [lemmatizer.lemmatize(word)
             for word in words if word not in stop_words]
    return " ".join(words)
```

```python
cleaned_texts = []

for text in texts:
    text = clean_text(text)
    text = preprocess(text)
    cleaned_texts.append(text)

print(cleaned_texts)
```

```python
#label encoding

le = LabelEncoder()
encoded_labels = le.fit_transform(labels)

print(encoded_labels)
```

```python
#TF-IDF Vectorization

tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(cleaned_texts)

df_tfidf = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=tfidf.get_feature_names_out()
)

print(df_tfidf)
```

```python
#save cleaned text
pd.DataFrame({
    "Cleaned_Text": cleaned_texts,
    "Encoded_Label": encoded_labels
}).to_csv("processed_text.csv", index=False)
```

```python
#TF-IDE OUTPUT Save
df_tfidf.to_csv("tfidf_output.csv", index=False)
```
